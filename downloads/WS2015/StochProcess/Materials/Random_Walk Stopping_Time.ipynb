```python
%matplotlib inline

import numpy as np  # this imports the NumPy library
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(context="talk")

days = 101 # time horizon

def random_walk(startprice):
    price = np.zeros(days)
    shock = np.round(np.random.rand(days))
    price[0] = startprice
    for i in range(1, days):
        price[i] = price[i-1] + 2*shock[i]-1
    return price


fig, axn = plt.subplots(figsize=(15, 15))
for run in range(4):
    axn.plot(random_walk(10), label='time', linewidth='2')

axn.plot([0, 101], [3, 3], color='k', linestyle='-', linewidth=3)
```
